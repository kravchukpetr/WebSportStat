```python
for key, value in country_dict.items():
#     if key in ['Украина']:
    response = req.get('https://www.sport-express.ru' + value['url'])
    print('https://www.sport-express.ru' + value['url'])
    soup_country = BeautifulSoup(response.text, 'lxml')
    league_dict = {}
    if value['main_country'] == 1:
        league_list = soup_country.find("ul", {"class": "se19-nav"})
        if league_list is not None:
            for option in league_list.find_all('a'):
                if option.text != 'Новости':
                    league_dict[option.text] = option['href']
    #                 print (f"value: {option['href']}, text: {option.text}")
    if value['main_country'] == 0:
        league_list = soup_country.find("select", {"id": "submenu_level_b_2"})
        if league_list is not None:
            for option in league_list.find_all('option'):
                league_dict[option.text] = "https://sport-express.ru" + option['value']
    #             print (f"value: {option['value']}, text: {option.text}")

    country_dict[key]['Leagues'] = league_dict
#print(soup_country.prettify())
```
